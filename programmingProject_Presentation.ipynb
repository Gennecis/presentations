<h1 style="color: indigo">Purpose of the application<h1/>
    <p style="font-size:25px">
1. Item Tracking: The program will let users input items as they shop. This way, they can keep a record of everything they purchase.<br><br>
2. Unit Calculation: For each item, the program will ask users to specify the quantity or units they are purchasing. This ensures accurate tracking and helps users understand how much of each item they've purchased.<br><br>
3. Total Spending Calculation: The program will add up the costs of all the items the user has entered. This feature provides an easy way for users to know the total amount they've spent during their shopping session even before they checkout.<p/>

<h1 style="color: indigo">Purpose of the application continued<h1/>
    <p style="font-size:25px">
4. Budget Awareness: By calculating the total spending, the program helps users see how their purchases align with their budget. This way, they can make sure they're not overspending and manage their finances better.<br><br>
5. Expense History: The program will maintain a history of past shopping sessions and their respective costs. This allows users to compare spending across different trips and make informed decisions about their future shopping.<br>
In essence, the program serves as a digital shopping receipt and expense tracker, helping users keep tabs on their spending and make wise financial choices.<p/>
    

<h1 style="color: indigo"> A break-down of what the application will do<h1/>
<p style="font-size:25px">
1. allow input of items<br>
2. display added items<br>
3. remove previously added items<br>
4. compute total amount of items purchased<br>
5. save an item to be purchased later<br>
6. display items saved for later purchase
